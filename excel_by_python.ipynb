```python
import openpyxl

#파일 관리 1

#새로운 엑셀 파일 생성
wb = openpyxl.Workbook()

#시트 활성화
ws = wb.active()

# Sheet 이름 변경
ws.title = '자동화로만든거'

#엑셀 저장
wb.save('자동화된엑셀.xlsx')
```

```python
#파일관리 2
import openpyxl

wb = openpyxl.Workbook()

#시트 생성
ws = wb.create_sheet('2030.01')

#시트 삭제
del wb['Sheet']

wb.save('자동화된엑셀2.xlsx')

print(wb.sheetnames)
```

```python
#파일관리 3
import openpyxl

save_path = '자동화된엑셀2.xlsx'

#기존 엑셀 파일 불러오기
openpyxl.load_workbook(save_path) 

#활성화된 시트 선택
ws = wb.active 

#데이터 추가 (1)
ws['A1'] = '날짜'

#데이터 추가 (2)
ws.cell(row=1, column=1, value = '2030-01-01')
ws.cell(row=1, column=2, value= '=C2*D2') #이렇게 셀 값을 이용해 대입 가능

#데이터 추가 (3)
ws.append(['2030-01-03', '기계식 키보드', 120000, 15, "=C3*D3"])

#데이터 수정
ws['C2'] = 40000
ws['D2'] = 40

#데이터 삭제
del ws['A3']

#엑셀 저장 
wb.save(save_path)
```
