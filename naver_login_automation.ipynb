```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
import pyperclip #import pyautogui

#자동 로그인 캡챠를 피하려면 pyperclip을 써야 하는데 로컬 컴퓨터에서 실행하는게 아니다보니 불가능

#pyautogui되면 pyautogui.hotkey("ctrl", "v")

url = "https://nid.naver.com/nidlogin.login?mode=form&url=https%3A%2F%2Fwwwnaver.com"
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome('chromedriver', options=chrome_options)

driver.implicitly_wait(10)
driver.maximize_window()
driver.get(url)

id = driver.find_element_by_css_selector('#id')
id.click()
pyperclip.copy('joonsuheojs')
id.send_keys(Keys.CONTROL, 'v')
time.sleep(2)

pw = driver.find_element_by_css_selector("#pw")
pw.click()
pyperclip.copy("khu02!SOFTWARE")
id.send_keys(Keys.CONTROL, 'v')
time.sleep(2)

login_btn = driver.find_element_by_css_selector('#log.login')
login_btn.click()
time.sleep(2)
```
