```python
import datetime
format = "%Y-%m-%d %H:%M:%S"
datetime_str = "2018-05-13 12:34:56"
datetime_dt = datetime.datetime.strptime(datetime_str, format)
print(type(datetime_dt))
print(datetime_dt)

#str -> datetime: datetime.datetime.strptime(string 변수, 포맷)
#                 datetime_dt.strftime(포맷) 
#포맷: '%Y-%m-%d %H 등 다 안 쓰고 일부만 써도 됨

#strptime: str 타입으로 된 날짜 데이터를 datetime 형태로 변환 함수
#strftime: datetime 형태의 데이터를 str 타입으로 변환하는 함수
```

**2.2.4 금융 시계열 데이터 분석에 유용한 판다스 함수**

```python
#shift 함수
#주가를 분석하는 경우 전일 주가 대비 당일 주가의 변화량 혹은 변화율을 계산해 모멘텀을 계산하거나 수익률을 계산하는데 응용할 수 있다
#shift 함수는 인덱스에 연결된 데이터를 일정 간격으로 이동시키는 함수다. 인덱스에서는 변화 없이 데이터만 전, 후로 이동한다; 밑으로 내림.
#shift 함수를 호출 할 때 아무 숫자도 전달하지 않으면 기본적으로 'period' 값에 1이 전달된다
#epriods 값에 정수형 데이터가 전달되면 숫자만큼 데이터가 이동한다. 음의 정수 데이터가 전달되면 위 방향으로 시프트되고 양의 정수 데이터가 전달되면 아래로 내려간다
#axis='1'(column) 값을 전달하면 오른쪽으로 이동하게 된다.

aapl_df['Close_lag1'] = aapl_df['Close'].shift()
aapl_df.head()
```

```python
#pct_change 함수
#현재 값과 이전 요소 값의 백분율 변화량을 연산하는 함수다
#periods 변수가 shift 함수와 똑같이 적용, 음의 정수 데이터가 전달되면 위 방향으로 변화율 계산 vice versa

aapl_df['pct_change'] = aapl_df['Close'].pct_change()
aapl_df.head(10)
```

```python
#diff 함수
#pct_change가 변화율을 계산한다면 diff 함수는 변화량을 계산한다
#아래에서 위로 뺌

aapl_df['Close_diff'] = aapl_df['Close'].diff()
aapl_df.head()
```

```python
#rolling 함수
#rolling 함수는 윈도우라는 일정 구간 데이터들의 평균값, 최솟값, 최댓값 등을 계산하는 함수다

df['MA'] = df['Close'].rolling(window = 5).mean()
df.head(10)
```

```python
#resample 함수
#시간 간격을 재조정하는 기능을 한다
#업 샘플링: 분 단위, 초 단위로 샘플의 빈도수를 증가시킨다
#다운 샘플링: 몇 일, 몇 달 단위로 샘플의 빈도수를 감소시킨다

import pandas as pd
index = pd.date_range(start='2019-01-01', end= '2019-10-01', freq='B')
series = pd.Series(range(len(index)), index=index)
series
```

```python
series.resample(rule='M').sum()
```

```python
#rule의 단위로는 날짜의 주기 단위와 같은 단위를 사용한다

series.resample(rule='M').last()
```

```python
series.resample(rule='MS').first()
```
